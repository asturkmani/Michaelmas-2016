```python
import cPickle as cp
import numpy as np
import matplotlib.pyplot as plt

#import white wine data
X, y = cp.load(open('winequality-white.cPickle','rb'))

#split data into training and testing data
N, D = X.shape
N_train = int(0.8*N)
N_test = N-N_train

X_train = X[:N_train]
y_train = y[:N_train]
X_test = X[N_train:]
y_test = y[N_train:]
```

```python
#Handin 1: Plot bar chart showing distribution of y values
plt.hist(y)
plt.title('Distribution of wine ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
#Handin 2: Using the mean as a simple, report the mean square error:

#Compute the mean of y values to use as a trivial predictor
ytr_mean = y_train.mean()

#Compute MSE when predictions are made on the test set
yte_predict = np.full((N_test,1), ytr_mean)
mse_trivial_test = ((yte_predict - y_test)**2).mean()

print("Mean Squared Error with trivial prediction on test data is: ", mse_trivial_test)

#Compute MSE when predictions are made on the train set
ytr_predict = np.full((N_train,1), ytr_mean)
mse_trivial_train = ((ytr_predict - y_train)**2).mean()

print("Mean Squared Error with trivial prediction on train data is: ", mse_trivial_train)
```

```python
#Handin 3: Linear model using least squares

#Standardize the data
X_mu = X_train.mean(axis=0)
X_std = X_train.std(axis=0)

X_train_s = (X_train - X_mu)/(X_std)
X_test_s = (X_test - X_mu)/(X_std)


#Append the bias column to train and test data
X_train_s = np.c_[np.ones((N_train,1)), X_train]
X_test_s = np.c_[np.ones((N_test,1)), X_test]

#Compute weights using closed form solution
w = np.dot(np.dot(np.linalg.inv(np.dot(X_train_s.T, X_train_s)),X_train_s.T),y_train)

#Compute MSE when predictions are made on the test set
yte_predict = np.dot(X_test_s,w)

mse_LS_test = ((yte_predict - y_test)**2).mean()
print("Mean Squared Error with LS prediction on test data is: ", mse_LS_test)

#Compute MSE when predictions are made on the trainig set
ytr_predict = np.dot(X_train_s,w)
mse_LS_train = ((ytr_predict - y_train)**2).mean()
print("Mean Squared Error with LS prediction on train data is: ", mse_LS_train)
```

```python
# Learning Curves

errors_train = np.zeros((30,1))
errors_test = np.zeros((30,1))

for i in xrange(20,620,20):
    
    #Select section of data set
    Xp_train = X_train[:i,:]
    yp_train = y_train[:i]
    
    #Standardize the data
    X_mu = Xp_train.mean(axis=0)
    X_std = Xp_train.std(axis=0)

    X_train_s = (Xp_train - X_mu)/(X_std)
    X_test_s = (X_test - X_mu)/(X_std)


    #Append the bias column to train and test data
    Xp_train_s = np.c_[np.ones((i,1)), Xp_train]
    X_test_s = np.c_[np.ones((N_test,1)), X_test]

    #Compute weights using closed form solution
    w = np.dot(np.dot(np.linalg.inv(np.dot(Xp_train_s.T, Xp_train_s)),Xp_train_s.T),yp_train)

    #Compute MSE when predictions are made on the test set
    yte_predict = np.dot(X_test_s,w)
    mse_LS_test = ((yte_predict - y_test)**2).mean()
    errors_test[i/20-1] = mse_LS_test

    #Compute MSE when predictions are made on the trainig set
    ytr_predict = np.dot(Xp_train_s,w)
    mse_LS_train = ((ytr_predict - yp_train)**2).mean()
    errors_train[i/20-1] = mse_LS_train

size=np.linspace(20,600,30)
plt.plot(size,errors_train,label='Train errors')
plt.plot(size,errors_test,label='Test errors')
plt.legend()
plt.show()
```

# Analysis

From the above plot, we can see that for data sizes less than 300, the model is underfitting the data. We have a relatively small data set as compared to the number of features which doesn't allow the model to capture the underlying trends in the data. Beyond the 300 datapoints level, however, we see the error rate leveling off to about 0.55, which seems to be the optimal error rate.

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import Ridge, Lasso
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV

#We have the sets X_train, y_train, X_test, y_test

# Separate training data into new training data and validation data.
# Procedure: Loop over all degrees
#                 Loop over all powers of lambda
#                      Create polynomial features
#                      Scale features

# calculate size of new training list
Nn_train = int(0.8*N_train)

#create training and validation lists
Xn_train = X_train[:Nn_train]
yn_train = y_train[:Nn_train]

X_validate = X_train[Nn_train:]
y_validate = y_train[Nn_train:]

powers = 10.0**np.linspace(-5,5,11, dtype=int)
degrees = np.array([2,3,4])

# poly = PolynomialFeatures(include_bias=False)
# scaler = StandardScaler()
# ridge = Ridge()
# lasso = Lasso()

scoreR = []
scoreL = []
```

```python
for deg in degrees:
    for power in powers:
        # we're not including a bias here so it's not centred to 0 in the next step, instead use bias in LASSO/Ridge
        poly = PolynomialFeatures(degree=deg, include_bias=False)
        scaler = StandardScaler()
        ridge = Ridge(alpha=power)
        lasso = Lasso(alpha=power)
        
        #create pipeline
        pipeR = make_pipeline(poly, scaler, ridge)
        pipeL = make_pipeline(poly, scaler, lasso)
        
        #fit the models
        pipeR.fit(Xn_train, yn_train)
        pipeL.fit(Xn_train, yn_train)
        
        #make predictions
        yR_predict = pipeR.predict(X_validate)
        yL_predict = pipeL.predict(X_validate)
        

        #score using mean squared error
        scoreR.append(((y_validate - yR_predict)**2).mean())
        scoreL.append(((y_validate - yL_predict)**2).mean())

print "minimum value of Ridge is: ", min(scoreR), "located at index ", scoreR.index(min(scoreR))
print "minimum value of Lasso is: ", min(scoreL), "located at index ", scoreL.index(min(scoreL))

```

```python
# Find the values that correspond to this maximum:

degOptimalR = degrees[int(scoreR.index(min(scoreR))/len(powers))]
powOptimalR = powers[int(scoreR.index(min(scoreR))%len(powers))]

print "optimal values of degree and lambda for Ridge are: ", "d:",degOptimalR, "L:",powOptimalR

degOptimalL = degrees[int(scoreL.index(min(scoreL))/len(powers))]
powOptimalL = powers[int(scoreL.index(min(scoreL))%len(powers))]

print "optimal values of degree and lambda for Lasso are: ", "d:", degOptimalL, "L:", powOptimalL
```
